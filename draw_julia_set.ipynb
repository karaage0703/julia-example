```julia
function cnt(z, c)
    k::UInt8 = 0
    while (k < 255)
        z = z^2 + c
        abs2(z) > 4. && return k
        k += 1
    end
    return k
end
```

```julia
function jlset(M::Int, N::Int, c)
    grid = Array{UInt8}(undef, M, N)
    xs = range(-2,2,length=N)
    ys = range(-2,2,length=M)
    grid = @. cnt(complex(xs',ys), c)
    return grid
end
```

```julia
c = complex(-0.38, 0.6)
M=N=500
heatmap(1:N, 1:M, jlset(M,N,c), aspect_ratio=:equal)
```
